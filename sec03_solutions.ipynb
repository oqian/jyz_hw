```python
import pandas as pd
import numpy as np
# Download the diabetes file, and save it in the working directory
# Read the text file into a pandas dataframe
diabetes_df = pd.read_csv("diabetes.txt", sep='\t')

#1. Y is the output variable for the dataframe. Y is a measure of how far the disease has progressed in 1 year after the initial data collection.

columns = set(diabetes_df.columns)
columns.remove("Y")
for column in columns:
    col_values = np.array(diabetes_df[column])
    # calculate the mean and std for each attribute
    print(column, "\n std:", np.std(col_values), "\n mean:", np.mean(col_values))
    # calculate the standard deviation
    np.std(col_values)
    
# calculate the correlation coefficient matrix
print("Correlation coefficient matrix (computed with numpy):") 
display(pd.DataFrame(np.corrcoef(np.array(diabetes_df).T),columns = diabetes_df.columns))

print("Correlation coefficient matrix (computed with pandas):")
display(diabetes_df.corr())
```

The main diagonal of the correlation coefficient is equal to a vector of ones because every variable has full correlation with itself. 

3. It does not truly make sense to calculate the mean or standard deviation for the SEX column. It is a categorical variable, and thus mean does not make sense. Furthermore, standard deviation is not useful, since there are no values besides 1 and 2.

###### Part 2: Use Linear Regression on Quadratic Features
Use sklearn.preprocessing.PolynomialFeatures to expand the data set to include nonlinear elements, then perform linear regression again. Is this a better model comparing with the one you find in part 1?

reference:https://scikit-learn.org/stable/modules/generated/sklearn.preprocessing.PolynomialFeatures.html

```python
from sklearn.preprocessing import PolynomialFeatures
poly_transform = PolynomialFeatures(2)
```

```python
# Sample the train and test sets
train, test = train_test_split(lin_diab_view, test_size=0.2)
train_X = train.drop(columns = ["Y"])
train_y = train["Y"]
test_X = test.drop(columns = ["Y"])
test_y = test["Y"]

# Transform lin_diab_view to have nonlinear interaction 
X = lin_diab_view.drop(columns = ["Y"])
poly_diab_view = poly_transform.fit(X)
p_train_X = poly_diab_view.transform(train_X)
p_test_X = poly_diab_view.transform(test_X)

p_diab_pred = LinearRegression().fit(p_train_X, train_y)

print("Testing for all polynomial expansions of the features for degree <= 2")
print(f"R2 values for the train set: {p_diab_pred.score(p_train_X,train_y)}")
print(f"MSE value for the train set: {mean_squared_error(train_y,p_diab_pred.predict(p_train_X))}")
print(f"R2 values for the test set: {p_diab_pred.score(p_test_X, test_y)}")
print(f"MSE value for the test set: {mean_squared_error(test_y,p_diab_pred.predict(p_test_X))}")

p_train_X, p_test_X, train_y, test_y = train_test_split(normalized_dataset, normalized_output, test_size=0.2)
p_diab_pred = LinearRegression().fit(p_train_X, train_y)
test_preds = p_diab_pred.predict(p_test_X)
train_preds = p_diab_pred.predict(p_train_X)

print("Testing for all polynomial expansions of the  normalized features for degree <= 2")
print(f"R2 values for the train set: {p_diab_pred.score(p_train_X,train_y)}")
print(f"MSE value for the train set: {mean_squared_error(y_scaler.inverse_transform(train_y.astype(float)),y_scaler.inverse_transform(train_preds.astype(float)))}")
print(f"R2 values for the test set: {p_diab_pred.score(p_test_X, test_y)}")
print(f"MSE value for the test set: {mean_squared_error(y_scaler.inverse_transform(test_y.astype(float)),y_scaler.inverse_transform(test_preds.astype(float)))}")
```

###### Notice that for the polynomial factor expansion in this case, the result varies widely with different samplings of the train and test set. In part, this is because we have such a small set with respect to the number of features. When you have many features and a small dataset, you run the risk of overfitting. 

###### Note as well, that the normalized model in this case does not show nearly as much instability
